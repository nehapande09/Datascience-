# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

Ans: The ArithmeticError class is a base class for exceptions that occur during arithmetic operations in Python. It is part of the Python exception hierarchy and serves as a parent class for several specific arithmetic-related exceptions. Two notable exceptions derived from the ArithmeticError class are FloatingPointError and ZeroDivisionError.

1. FloatingPointError:
FloatingPointError is raised when an error occurs during a floating-point arithmetic operation.
2. ZeroDivisionError:
ZeroDivisionError is raised when you attempt to divide a number by zero. It is one of the most common arithmetic exceptions in Python and occurs when the denominator in a division operation is zero.



```python
# Floating Point Error
try:
    result = 1.0 / 0.0
except ArithmeticError as e:
    logging.info(f"FloatingPointError: {e}")
```

```python
# ZeroDivision error
try:
    1/0
except ZeroDivisionError as e:
    logging.info(f"ZeroDivision error : {e}")
```

# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Ans:
    The LookupError class in Python is a base class for exceptions that represent errors related to looking up or indexing items in sequences, mappings, or other iterable objects. It serves as a parent class for specific lookup-related exceptions. Two common exceptions derived from the LookupError class are KeyError and IndexError.
    
   1. KeyError is raised when you attempt to access a dictionary using a key that does not exist in the dictionary. It is specific to dictionary data structures in Python.
   2. IndexError is raised when you try to access an index in a sequence or a list that is outside the valid range of indices (i.e., the index is either too large or negative)

```python
# keyError
my_dict = {'a': 1, 'b': 2, 'c': 3}
try:
    value = my_dict['x']  
except KeyError as e:
    logging.info(f"KeyError: {e}")
```

```python
#IndexError
my_list = [1, 2, 3]
try:
    value = my_list[5]  
except IndexError as e:
    logging.info(f"IndexError: {e}")
```

# Q5. Explain ImportError. What is ModuleNotFoundError?

Ans:
      ImportError and ModuleNotFoundError are both exceptions in Python related to importing modules. These exceptions occur when there is an issue with importing a module or package.

1. ImportError:
ImportError is a base class for exceptions related to importing modules or packages. It is raised when there is an issue with the import statement. ImportError can have various subtypes depending on the specific problem, such as missing modules, circular imports, or other import-related errors.

2. ModuleNotFoundError:
ModuleNotFoundError is a more specific exception that was introduced in Python 3.6. It is derived from ImportError and is raised when Python cannot find the specified module. This exception is more descriptive and provides clearer information about the missing module.


```python
# ImportError
try :
    import nomodule
except ImportError as e:
    logging.info(f"Import Error : {e}")
```

```python
# Module Not Found error
try :
    import neha
except ModuleNotFoundError as e:
    logging.info(f"Module Not Found Error: {e} ")
```
